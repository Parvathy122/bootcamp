```python
labels = list(df.Entity.value_counts().to_dict().keys())
Afghanistan = list(df.Entity.value_counts().to_dict().values())[0]
Guam = list(df.Entity.value_counts().to_dict().values())[1]
```

```python
labels
```

```python
plt.figure(figsize=(7,7))
plt.title("Entity v/s Year")

plt.bar(df[df['Entity']== 'Afghanistan'].Entity.value_counts().to_dict().keys(),
        df[df['Entity']== 'Afghanistan'].Entity.value_counts().to_dict().values());

plt.bar(df[df['Entity']== 'Guam'].Entity.value_counts().to_dict().keys(),
        df[df['Entity']== 'Guam'].Entity.value_counts().to_dict().values());

plt.xlabel('Entity')
plt.xlabel("Year");
```

```python
plt.figure(figsize=(7,7))
plt.title("Entity v/s Alochol use ")
sns.barplot(x='Entity',
            y='Alochol use',
            palette = "viridis",
            data=df);
```
